### Extract Dataset # 1 GPT

```python
# Extract the GPT responses
gpt_responses = [row['row']['conversations'][1]['value'] for row in file['rows']]

# # Print the GPT responses
# for response in gpt_responses:
#     print(response)
```

```python
# remove newlines and back slashes and commas
v1 = []

for text in gpt_responses:
    cleaned_text = re.sub(r'[\n\\]', '', text)
    v1.append(cleaned_text)

v1

```

```python
# remove the coding isntructions 
#v1[1]
v1_1 = v1[:1] + v1[1 + 1:]
```

```python
# label and save words list
word_list = [word.lower() for paragraph in v1_1 for word in paragraph.split()]

# join all strings in v1_1 into one string
one_string = " ".join(v1_1)
```

```python
# POS tag the words
import nltk
from nltk import *



clean_gpt = []
try:
  for word in v1_1:

    # process sentences
    cleane = preprocess(word)
    clean_gpt.append(cleane)

except KeyError:
  print("Column 'Data' not found")
```

```python
clean_gpt
```

```python
# bi,tri,quad phrases after removing porper nouns
QT_gpt =[]
for pair in clean_gpt:

    for word, tag in pair:
        if tag != "NNP" and tag != ":" and tag != "''" and tag != "``" and tag != ".":
            QT_gpt.append(word.lower())

        elif tag == "NNP": 
            q = "Q"
            QT_gpt.append(q)

        elif tag == ":":
            q = "T"   # T for punct
            QT_gpt.append(q)

        elif tag == ".":
            q = "T" # T for punctuation 
            QT_gpt.append(q)
```

```python
# save QT list 
df = pd.Series(QT_gpt)
df.to_csv('QT dataset 1.csv', index = False)
```
