```python
import matplotlib as mpl
import matplotlib.patches as patches
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Admin\HC3 EDA\GriD\all_words4.csv")
```

```python

# Create a matplotlib table
plt.figure(figsize=(12, 6))
table = plt.table(cellText=df.values,
                  colLabels=df.columns,
                  cellLoc='center',
                  loc='upper left')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

# Remove axis
plt.axis('off')

# Add title
plt.title("Word Count Statistics", fontsize=16, fontweight='bold')

# Show the table
plt.show()
```

```python
# Create a matplotlib table with the test data
plt.figure(figsize=(8, 4))
table = plt.table(cellText=df.values,
                  colLabels=df.columns,
                  cellLoc='center',
                  loc='upper left')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

# Remove axis
plt.axis('off')

# Calculate the figure size and the number of rows in the table
fig_size = plt.gcf().get_size_inches()
num_rows = len(df)

# Calculate the height of the figure in pixels
fig_height_pixels = fig_size[1] * plt.gcf().dpi

# Calculate the approximate number of pixels per row
pixels_per_row = fig_height_pixels / num_rows

print(f"Approximate number of pixels per row: {pixels_per_row:.2f}")

# Show the table
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Population (millions)': [8.4, 3.9, 2.7, 2.3, 1.7],
    'Area (sq miles)': [468.9, 469.3, 227.3, 637.5, 517.6]
}
df = pd.DataFrame(data)

# Convert DataFrame to a list of lists
data_values = [df.columns.tolist()] + df.values.tolist()

# Create a matplotlib table
plt.figure(figsize=(8, 4))
table = plt.table(cellText=data_values,
                  colLabels=df.columns,
                  cellLoc='center',
                  loc='upper left')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

# Remove axis
plt.axis('off')

# Add title
plt.title("City Statistics", fontsize=16, fontweight='bold')

# Show the table
plt.show()
```
